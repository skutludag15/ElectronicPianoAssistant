```python
import cv2
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

img = cv2.imread("aug-gutenberg1939--page-2.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite("lol.png",gray)
ret,thresh2 = cv2.threshold(gray,200,255,cv2.THRESH_BINARY_INV)
cv2.imwrite("lol1.png",thresh2)

horizontal = np.copy(thresh2)
cols = horizontal.shape[1]
horizontal_size = cols // 30
horizontalStructure = cv2.getStructuringElement(cv2.MORPH_RECT, (horizontal_size, 1))
horizontal = cv2.erode(horizontal, horizontalStructure)
horizontal = cv2.dilate(horizontal, horizontalStructure)
cv2.imwrite("lol2.png",horizontal)

count = [None for _ in range(0, horizontal.shape[0])]
for y in range(0, horizontal.shape[0]): #looping through each rows
    count[y] = 0
    for x in range(0, horizontal.shape[1]): #looping through each column
        if horizontal[y, x] > 250:
            count[y] = count[y] + 1

staff = 0
counter = 0
list = [[None,None] for _ in range(0, 100)]
for y in range(0, horizontal.shape[0]):
    if(count[y] > horizontal.shape[0] / 3):
        list[counter] = [staff,y]
        counter = counter + 1
        if(counter > 1):
            if(y - list[counter-2][1] > 10):
                staff = staff + 1
                list[counter-1] = [staff,y]
print(list)                
counter = 1
staff = 0 
answer = 0
list2 = []

for y in range(0,len(list)):
    if(list[y][0] == staff):
        answer = answer + list[y][1]
        counter = counter + 1
    elif(list[y][0] == staff + 1):
        answer = answer / (counter - 1)
        list2.append(answer)
        answer = 0
        counter = 1
        staff = staff + 1
    elif(list[y][0] == None):
        answer = answer / (counter - 1)
        list2.append(answer)
        answer = 0
        counter = 1
        staff = staff + 1
        break
    
        
print(list2)               
               
```
